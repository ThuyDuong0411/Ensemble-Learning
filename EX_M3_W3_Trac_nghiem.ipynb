```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor

data = {'X': [3,5,8,10,12],
        'Y': [12,20,28,32,36]}

# load data
df = pd.DataFrame(data)
df
```

```python
# get value and convert to numpy
X = df[['X']].values
y = df[['Y']].values.reshape(-1,)

# define regressor
reg = DecisionTreeRegressor()

# train
reg = reg.fit(X, y)
```

```python
# predict
x_test = np.array([[15]])
predicted_label = reg.predict(x_test)
predicted_label
```

```python
from sklearn.tree import plot_tree

# visualization
plot_tree(reg, feature_names=['X'], fontsize=10)
```

```python
data = {'X': [2,1,3,2],
        'Y': [4,3,5,6]}

# load data
df = pd.DataFrame(data)
df
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Get data
x_data = df[['X']].to_numpy()
x_data = x_data.reshape(4, 1)
y_data = df['Y'].to_numpy(dtype=np.uint8)

#Define model
rf_regressor = RandomForestRegressor(
    n_estimators=2,
    max_depth=1,
    max_samples=4
)

# Train model
rf_regressor.fit(x_data, y_data)

# Predict
x_test = np.array([[2]])
y_pred = rf_regressor.predict(x_test)
y_pred
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
_ = tree.plot_tree(
    rf_regressor.estimators_[0],
    feature_names=['X'],
    filled=True
)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
_ = tree.plot_tree(
    rf_regressor.estimators_[1],
    feature_names=['X'],
    filled=True
)
```
